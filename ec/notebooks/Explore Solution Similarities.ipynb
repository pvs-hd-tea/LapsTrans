```python
import os.path as osp
import json 

checkpoint = '../experimentOutputs/logo/2020-02-16T21:46:57.648895/logo_aic=1.0_arity=3_aux=True_BO=True_CO=True_ES=1_ET=3600_HR=0.5_it=20_MF=5_noConsolidation=False_pc=30.0_RS=5000_RT=3600_RR=False_RW=False_solver=ocaml_STM=True_L=1.5_batch=50_TRR=randomShuffle_K=2_topkNotMAP=False_graph=True.pickle'
outputDir = osp.join(osp.dirname(checkpoint), 'solutions')

with open(osp.join(outputDir, 'primitive_counts.json'), 'r') as f:
    count_data = json.load(f)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(10, 100))

# Final DSL
unigrams = count_data['final_dsl']['unigram_uses']
frontier_names = sorted(count_data['final_dsl']['frontiers_to_idx'].keys(), key=lambda k : count_data['final_dsl']['frontiers_to_idx'][k])
Z = linkage(unigrams, 'ward')
dn = dendrogram(Z, labels = frontier_names, orientation='left')
plt.show()
```

```python
# Final DSL
fig = plt.figure(figsize=(25, 10))
bigrams = [b[0] for b in count_data['final_dsl']['bigram_uses']]
print(len(bigrams[0]))
frontier_names = sorted(count_data['final_dsl']['frontiers_to_idx'].keys(), key=lambda k : count_data['final_dsl']['frontiers_to_idx'][k])
Z = linkage(bigrams, 'ward')
dn = dendrogram(Z, labels = frontier_names)
plt.show()
```
